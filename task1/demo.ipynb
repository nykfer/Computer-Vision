# Simple demo for Task 1
#
# This notebook shows how to:
# 1. Load MNIST using torchvision
# 2. Train one of the classifiers (rf, nn, cnn)
# 3. Get predictions wrapped in MNISTPredict
# 4. Visualize a few predicted digits


```python
# Imports
import torch
from torchvision import datasets, transforms

from task1.mnist_classifier import MnistClassifier  # unified interface

# Check device (we keep training very small so CPU is fine too)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Models will handle data preprocessing (flattening/normalization) internally.
```

```python
# Load a small subset of MNIST
# We keep the subset small so that the demo runs quickly.
#
# NOTE: Models now handle flattening and normalization internally,
# so we can pass raw image tensors from the dataset.

transform = transforms.ToTensor()

train_dataset = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root="./data", train=False, download=True, transform=transform)

# Take a small subset for quick training (e.g. first 2000 train, 500 test samples)
train_size = 2000
test_size = 500

X_train = train_dataset.data[:train_size]
y_train = train_dataset.targets[:train_size]

X_test = test_dataset.data[:test_size]
y_test = test_dataset.targets[:test_size]

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```

```python
# Initialize and train the classifier
# Possible algorithms: "rf" (RandomForest), "nn" (feed-forward NN), "cnn" (convolutional NN)

algorithm = "cnn"  # change to "rf" or "nn" to try other models
classifier = MnistClassifier(algorithm=algorithm)

# Train on the small subset
classifier.train(X_train, y_train)

print(f"Training finished for algorithm = {algorithm}")
```

```python
# Evaluate on the test subset and visualize some predictions

mnist_pred = classifier.predict(X_test, y_test)

# Overall accuracy (computed from predictions vs. ground truth)
import numpy as np

pred_labels = np.array(mnist_pred.preds)
true_labels = y_test.numpy()
accuracy = (pred_labels == true_labels).mean()
print(f"Overall accuracy on {len(true_labels)} test samples: {accuracy:.3f}")

# Per-class accuracy (if available)
if mnist_pred.accuracy is not None:
    print("Per-class accuracy (classes 0-9):")
    for cls, acc in enumerate(mnist_pred.accuracy):
        print(f"Class {cls}: {acc:.3f}")

# Show a small grid of predicted digits (images come from MNISTPredict.data)
mnist_pred.show(count=9)
```
